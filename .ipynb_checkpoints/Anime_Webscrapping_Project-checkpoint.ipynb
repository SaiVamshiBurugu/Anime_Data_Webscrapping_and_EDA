## 1. Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import time
from bs4 import BeautifulSoup
import requests
import random
import warnings
warnings.filterwarnings('ignore')
```

## 2. Web Scrapping

```python
headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                          "AppleWebKit/537.36 (KHTML, like Gecko) "
                          "Chrome/115.0 Safari/537.36"
        }
```

#### Example of the website

```python
title_list = []
rating_list = []
studio_list = []
source_list = []
release_year_list = []
genre_list = []
duration_list = []
no_of_episodes_list = []
type_of_show_list = []
description_list = []
```

```python
def extract_info(container, tag, class_name, attr_name=None):
    element = container.find(tag, class_=class_name)
    if not element:
        return np.nan
    
    if attr_name:  # If attribute is specified
        return element.get(attr_name, np.nan)
    else:  # Otherwise get text
        return element.get_text(strip=True)
```

```python
for i in range(1,44,2):
    url = f'https://www.anisearch.com/anime/toplist/page-{i}?char=all&type=0&view=1&limit=100&hentai=no'
    response = requests.get(url,headers = headers)
    print(f"Scrapping the URL : {url}","&",f"Response Code:{response.status_code}")
    soup = BeautifulSoup(response.text,'html.parser')
    containers = soup.find_all('li',class_='btype0')
    for container in containers:
        title_list.append(extract_info(container,'span','title'))
        rating_list.append(extract_info(container,'div','star0','title'))
        studio_list.append(extract_info(container,'span','company'))
        source_list.append(extract_info(container,'div','type'))
        genre_list.append(extract_info(container,'div','genre'))
        duration_list.append(extract_info(container,'div','episodes'))
        description_list.append(extract_info(container,'div','text scrollbox'))
        element = extract_info(container,'span','date')
        if element:
            pattern = r'([a-zA-Z\-\s]+)(?:,\s*(\d+|\?)\s*)?\((?:(\d{4})|(\?))\)'
            re_element = re.findall(pattern,element)
            type_of_show_list.append(re_element[0][0])
            no_of_episodes_list.append(re_element[0][1])
            release_year_list.append(re_element[0][2])
        else:
            type_of_show_list.append(np.nan)
            no_of_episodes_list.append(np.nan)
            release_year_list.append(np.nan)
    time.sleep(random.uniform(3, 7))
print("Scrapping Ended")
```

```python
for i in range(2,45,2):
    url = f'https://www.anisearch.com/anime/toplist/page-{i}?char=all&type=0&view=1&limit=100&hentai=no'
    response = requests.get(url,headers = headers)
    print(f"Scrapping the URL : {url}","&",f"Response Code:{response.status_code}")
    soup = BeautifulSoup(response.text,'html.parser')
    containers = soup.find_all('li',class_='btype0')
    for container in containers:
        title_list.append(extract_info(container,'span','title'))
        rating_list.append(extract_info(container,'div','star0','title'))
        studio_list.append(extract_info(container,'span','company'))
        source_list.append(extract_info(container,'div','type'))
        genre_list.append(extract_info(container,'div','genre'))
        duration_list.append(extract_info(container,'div','episodes'))
        description_list.append(extract_info(container,'div','text scrollbox'))
        element = extract_info(container,'span','date')
        if element:
            pattern = r'([a-zA-Z\-\s]+)(?:,\s*(\d+|\?)\s*)?\((?:(\d{4})|(\?))\)'
            re_element = re.findall(pattern,element)
            type_of_show_list.append(re_element[0][0])
            no_of_episodes_list.append(re_element[0][1])
            release_year_list.append(re_element[0][2])
        else:
            type_of_show_list.append(np.nan)
            no_of_episodes_list.append(np.nan)
            release_year_list.append(np.nan)
    time.sleep(random.uniform(3, 7))
print("Scrapping Ended")
```

```python
cols = [title_list,rating_list,studio_list,source_list,release_year_list,genre_list,duration_list,no_of_episodes_list,type_of_show_list,description_list]
for col in cols:
    print(len(col))
```

## 3. DataFrame Creation

```python
df = pd.DataFrame({
    'Title':title_list,
    'Rating':rating_list,
    'Studio':studio_list,
    'Adaptation':source_list,
    'Genre':genre_list,
    'Duration':duration_list,
    'Show Type':type_of_show_list,
    'No of Episodes':no_of_episodes_list,
    'Year of Release':release_year_list
})
```

## 6. Data Cleaning

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.replace(r'^\s*\?\s*$', np.nan, regex=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.sample(3)
```

```python
df.columns
```

```python
df['Year of Release'] = df['Year of Release'].astype(int)
```

```python
df['Show Type'] = df['Show Type'].astype('category')
df['Genre'] = df['Genre'].astype('category')
df['Adaptation'] = df['Adaptation'].astype('category')
```

```python
df.sample()
```

```python
df['Numeric Rating'] = df['Rating'].str.extract(r'([\d.]+)').astype(float)
df['Verbal Rating'] = df['Rating'].str.extract(r'[0-9.]+\s*(.*)')
df['Verbal Rating'] = df['Verbal Rating'].astype('category')
df['Studio'] = df['Studio'].astype('category')
df['Duration'] = df['Duration'].str.extract(r'(\d+)').astype(float)
```

```python
df.drop(['Rating','Show Type'],axis = 1,inplace=True)
```

```python
df.dropna(inplace=True)
df.info()
```

### High Rated Studios(Top 10)

```python
studio_avg = (
    df.groupby("Studio")
    .agg(avg_rating=("Numeric Rating", "mean"),
         count=("Title", "count"))
    .reset_index()
)

studio_avg = studio_avg[studio_avg["count"] >= 20]
studio_stats = studio_avg.sort_values("avg_rating", ascending=False).head(10)

plt.figure(figsize=(8,3))
plt.barh(studio_stats["Studio"], studio_stats["avg_rating"], color='skyblue', edgecolor='darkblue')
plt.xlabel("Average Rating")
plt.ylabel("Studio")
plt.title("Top Studios by Average Rating")
plt.gca().invert_yaxis()   # highest rating on top
plt.show()
```

```python
df.groupby(by = 'Genre',observed=False)['Numeric Rating'].mean().reset_index().sort_values('Numeric Rating',ascending=False).reset_index().head(10)

genre_avg = (
    df.groupby("Genre")
    .agg(avg_rating=("Numeric Rating", "mean"),
         count=("Title", "count"))
    .reset_index()
)

genre_avg = genre_avg[genre_avg["count"] >= 20]
genre_stats = genre_avg.sort_values("avg_rating", ascending=False).head(10)
print(genre_stats)
plt.figure(figsize=(8,3))
plt.barh(genre_stats["Genre"], genre_stats["avg_rating"], color='skyblue', edgecolor='darkblue')
plt.xlabel("Average Rating")
plt.ylabel("Genre")
plt.title("Top Genres by Average Rating")
plt.gca().invert_yaxis() 
plt.show()
```

### Multivariate Analysis(HeapMap)

```python
top_genres = df['Genre'].value_counts().nlargest(10).index
df_top_genres = df[df['Genre'].isin(top_genres)]
df_top_genres['Genre'] = df_top_genres['Genre'].cat.remove_unused_categories()
df_top_genres['Studio'] = df_top_genres['Studio'].cat.remove_unused_categories()
```

```python
genre_source = df_top_genres.pivot_table(
    index="Genre", 
    columns="Adaptation", 
    values="Numeric Rating", 
    aggfunc="mean"
)

plt.figure(figsize=(12,8))
sns.heatmap(genre_source, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Average Ratings by Genre & Adaptation (Top 10 Genres)")
plt.show()
```

```python
top_studios = (
    df_top_genres['Studio'].value_counts()
    .head(10)
    .index
)

df_filtered = df_top_genres[df_top_genres['Studio'].isin(top_studios)]
genre_source = df_filtered.pivot_table(
    index="Genre", 
    columns="Studio", 
    values="Numeric Rating", 
    aggfunc="mean"
)

plt.figure(figsize=(12,8))
sns.heatmap(genre_source, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Average Ratings by Genre & Top 10 Studios")
plt.ylabel("Genre")
plt.xlabel("Studio")
plt.show()
```

## 8. Insights from Anime Dataset

- All numeric columns show **outliers**.
- From the pair plot we can observe:
  - Most anime episodes are **20–30 minutes long**.
  - Most anime series have **less than 200 episodes**.
  - Recently released anime tend to have **higher ratings**, which is also confirmed in the trend analysis.
- **Excellent-rated** anime are the most common, while **Masterpiece-rated** ones are very rare.
- **J.C. Staff** produced the highest number of anime, followed by **Toei Animation** and **Sunrise**.
- The most frequent genres are **Adventure**, **Comedy**, and **Action**.
- Most anime are adapted from **Manga**, followed by **Original works** and **Light novels**.
- The **average rating** of anime has **increased over time**.
- The **average episode duration** has stayed roughly the same over the years.
- Anime with **large episode counts** have decreased over time, with modern anime usually having **shorter seasons**.
- The **Adventure genre** started to grow significantly around **1992**.
- The top-rated studios are:
  1. **CloverWorks**
  2. **Kyoto Animation**
  3. **White Fox**
- The top-rated genres are:
  1. **Slice of Life Comedy**
  2. **Adventure Comedy**
  3. **Thriller**
- **Fighting Shounen** anime tend to have more episodes on average.
- Only **Rating** and **Year of Release** show a noticeable (weak) **positive correlation**.
- All **Masterpiece-rated anime** have episode durations between **20–30 minutes**.
- **Highest-rated anime:** *Fullmetal Alchemist: Brotherhood*
- **Lowest-rated anime:** *Glass no Kamen Desu ga*, which is also the **shortest anime** with only **3 episodes**.
- **Longest anime:** *Sazae-san* with an incredible **3000+ episodes**.
- Webtoon adaptations combined with the Adventure genre show higher ratings.
- Light Novel or Manhwa adaptations in the Drama genre also receive good ratings.
